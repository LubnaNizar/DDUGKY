## creating dataframe from list and list of dictionaries

```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

#### creating lists,which will be the columns

```python
name=['lubna','keerthi','nida','merlin','swathi','jobi']
age=[22,25,21,27,25,21]
```

```python
data=pd.DataFrame(list(zip(name,age)))
data
```

### specify the column name

```python
data=pd.DataFrame(list(zip(name,age)),columns=['NAME','AGE'])
data
```

#### creating a DataFramenfrom a list of dictionaries

```python
list_of_dictionary=[
    {'Name':'Lubna','Age':22},
    {'Name':'Nida','Age':21},
    {'Name':'Keerthi','Age':25},
    {'Name':'Merlin','Age':27},
    {'Name':'Swathi','Age':25},
    {'Name':'jobi','Age':21},
 ]
```

```python
data_dict=pd.DataFrame(list_of_dictionary)
data_dict
```

### HANDLING NULL VALUES

```python
data=pd.read_csv('Windspeed.csv')
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

### FILLING A COMMON VALUE TO ALL MISSING DATA

#### 1.FILLING 0 TO ALL MISSING VALUE

```python
data.fillna(0)
```

#### 2. ADDING MISSING DATA TOMINDIVIDUAL COLUMNS

```python
data.fillna({
    "Temperature":0,
    "Windspeed":6,
    "Status":"sunny"
})
```

### 3.FORWARD FILL(R0W)

```python
data.fillna(method='ffill')
```

### 4.BACKWARDFILL(ROW)

```python
data.fillna(method="bfill")
```

### 5.LIMITING THE FORWARD/BACKWARD FILL

```python
data.fillna(method="ffill",limit=2)
```

### 6.FILLING WITH PANDAS OBJECT

```python
data.fillna(data.mean())
```

### 7. FILLING FOR SPECIFIC RANGE OF COLUMNS

```python
data.fillna(data.mean()['Temperature':'Windspeed'])
```

### 8.INTERPOLATE MISSING VALUE

```python
data.interpolate()
```

### 9.TIME INTERPOLATE

```python
data.interpolate(limit=1,limit_direction='backward')
```
