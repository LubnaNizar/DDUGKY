## assign 

```python
import numpy as np
import pandas as pd
df=pd.read_csv('netflix_titles_2021.csv')
df.head()
```

```python
df.assign(after_2020=df['release_year']>2020)
```

 ## query 

```python
df.query('release_year == 2015 & type=="Movie"')
```

##  dropna

```python
df=df.dropna(subset=['country'])
df.head()
```

## drop

```python
df2=df.drop(columns=['cast','description'])
df2.head()
```

## pivot table

```python
pivot=pd.pivot_table(columns='rating',values='release_year',data=df2,aggfunc='median')
(pivot)
```
